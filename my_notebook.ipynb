```python
# Paso 1: Importar las librerías necesarias
import matplotlib.pyplot as plt
import numpy as np

# Paso 2: Generar datos aleatorios para el ejemplo
x = np.random.randn(50)  # 50 puntos aleatorios para el eje X
y = np.random.randn(50)  # 50 puntos aleatorios para el eje Y

# Paso 3: Crear el gráfico de dispersión
plt.scatter(x, y)

# Agregar títulos y etiquetas
plt.title('Gráfico de Dispersión Simple')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')

# Mostrar el gráfico
plt.show()
```
